# 00 Collect Google Trend Data
For catching search queries that are associated with crypto trading, I'm using the library [PyTrends](https://pypi.org/project/pytrends/) for scraping the Google Trends API. I will store the results into a CSV file for later use.

```python
from pytrends.request import TrendReq
from pytrends import dailydata

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["bitcoin", "cryptocurrency", "ethereum", "trading", "etf"]
pytrends.build_payload(kw_list, cat=0, geo='', gprop='')
```

```python
df = pytrends.get_historical_interest(kw_list, 
                             year_start=2015, 
                             month_start=1, 
                             day_start=1, 
                             hour_start=0, 
                             year_end=2020, 
                             month_end=3, 
                             day_end=10, 
                             hour_end=0, 
                             cat=0, 
                             geo='', 
                             gprop='', 
                             sleep=5)
```

```python
df
```

As we can see in the Plot, the Trend data is very volatile and has a high changing frequency. We will process it later on Data Preprocessing for Modelling:

```python
df.to_csv("data/trends_bitcoin_cc_eth_trading_etf.csv")
```
